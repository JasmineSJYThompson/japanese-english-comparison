```python
import pandas as pd

import os
print(os.getcwd())
```

```python
japanese_10k = pd.read_csv("core10K.csv")
print(japanese_10k.shape)
japanese_10k.head()
```

```python
count_1w = pd.read_csv("../data/japanese-english-comparision/count_1w.csv", header=None, names=["word", "count"])
print(count_1w.shape)
count_1w["rank"] = count_1w.index + 1
count_1w.head()
```

```python
jap_eng_words = japanese_10k["Definition"].tolist()
eng_words = count_1w["word"].tolist()
```

```python
common_words = []
for word in eng_words:
    if word in jap_eng_words:
        common_words.append(word)
print(len(common_words))
```

```python
common_words_df = pd.DataFrame(common_words)
common_words_df.to_csv("../data/japanese-english-comparision/jap_eng_common_10K.csv")
```

```python
common_words_df.head()
```

```python
common_words_df.rename(columns={0: "word"}, inplace=True)
common_words_df.index.rename("rank", inplace=True)
common_words_df.reset_index(inplace=True)
common_words_df.head()
```

```python
common_words_df.to_csv("../data/japanese-english-comparision/jap_eng_common.csv")
```

```python
common_words_df.head()
```

```python
df=pd.merge(japanese_10k,count_1w, left_on='Definition', right_on='word')
```

```python
df.head()
```

```python
df.sort_values(["rank", "#"], inplace=True)
df.reset_index(inplace=True)
df.head()
```

```python
df["overall_rank"] = df.index+1
```

```python
df.head()
```

```python
df.rename(columns={
    "Kanji":"kanji", "Reading":"reading", "Part":"word_type_japanese",
    "rank":"rank_english", "count": "count_english",
    "#":"rank_japanese", "Sentence (hide)": "sentence", "Translation": "translation"}, inplace=True)
df.head()
```

```python
df.drop(columns=["index", "Unnamed: 0", "Definition", "Jisho"], inplace=True)
```

```python
df.head()
```

```python
df.drop(columns=["translation"], inplace=True)
```

```python
print(df.columns.tolist())
```

```python
df2 = df2[["word", "kanji", "reading", "overall_rank", "rank_english", "rank_japanese",
           "count_english", "word_type_japanese"]]
```
